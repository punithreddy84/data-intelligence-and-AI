```python
disct ={
    "name1": "punith",
    'age' :19,
    'lk' :32
}
disct1={
    "name1" : "pawan",
    'age' : 20,
    'lk' :23
}
result = disct.copy()
for key, value in disct1.items():
    result[key]=result.get(key,0)+value
print(result)    
```
