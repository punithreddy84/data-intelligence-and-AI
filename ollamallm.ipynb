```python
import ollama 
```

```python


messages = []

while True:
    prompt = input("You: ")

    if prompt.lower() == "exit":
        break

    messages.append(
        {"role": "user", "content": prompt}
    )

    response = ollama.chat(
        model="qwen2.5:3b",
        messages=messages
    )

    reply = response["message"]["content"]

    print("Bot:", reply)

    messages.append(
        {"role": "assistant", "content": reply}
    )
```
